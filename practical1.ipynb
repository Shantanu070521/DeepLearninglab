```python
import numpy as np
```

```python
#vector addition
```

```python
#define
vec_a=np.array([1,2,3])
vec_b=np.array([4,5,6])
```

```python
#perform element wise additon
vec_c=vec_a+vec_b
print("---Vector Addition---")
print("Vector a:",vec_a)
print("Vector b:",vec_b)
print("Vector c:",vec_c)
```

```python
#matrix multiplication
matrix_x=np.array([[1,2,3],[4,5,6]])
matrix_y=np.array([[5,6],[7,8],[2,3]])
matrix_product=np.dot(matrix_x,matrix_y)
print("---Matrix Multiplication---")
print("Matrix x:\n",matrix_x)
print("Matrix y:\n",matrix_y)
print("Matrix product:\n",matrix_product)
```
